```python
arr = ["cat", "apple", "bat", "dog", "banana"]
f = {}
for word in arr :
    length = len(word)
    if length not in f :
        f[length] = []
    f[length].append(word)
    
f
```

```python
arr = [2,5,1,2,7,5,8,2]

seen = set()
duplicate = []

for i in arr:
    if i not in seen:
        seen.add(i)
    elif i not in duplicate:
        duplicate.append(i)
# print(seen)
print(duplicate)
```

```python
def flatten(num ) :
    n = []
    for i in num :
        if isinstance(i , list):
            n.extend(flatten(i))
            
        else :
            n.append(i)
    return(n)

nested = [1, [2, [3, 4], 5], [6]]

print(flatten(nested))
```
