## Balanced Brackets: Given a string containing (), {}, and [], determine whether the brackets are balanced and properly nested.

```python
text = "{[()]}"

stack = []

pairs = {
    ')': '(',
    '}': '{',
    ']': '['
}

balanced = True

for char in text:

    if char in "({[":
        stack.append(char)

    else:
        if not stack:
            balanced = False
            break

        top = stack.pop()

        if top != pairs[char]:
            balanced = False
            break

if stack:
    balanced = False

print(balanced)
```

5. Find the Maximum Sum Subarray

Given:

arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

Find the contiguous subarray with the maximum sum.

Expected output:

Maximum Sum = 6

Because:

[4, -1, 2, 1]

has sum 6.

This is a famous interview problem called Kadane's Algorithm.

```python
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]

current_sum = arr[0]
max_sum = arr[0]

for num in arr[1:]:
    current_sum = max(num, current_sum + num)
    max_sum = max(max_sum, current_sum)

print(max_sum)
```
